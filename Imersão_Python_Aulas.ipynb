```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Cópia de [Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Cópia de [Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Cópia de [Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_bing = pd.read_excel("/content/Cópia de [Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Bing")
df_bing
```

```python
df_principal = df_principal[['Ativo',  'Data', 	'Último (R$)', 	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)' : 'valor_final', 'Var. Dia (%)' : 'var_dia_pct' }).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['var_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] =  (df_principal['valor_final'] - df_principal['var_inicial'])* df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = ' {:.2f}'.format
```

```python
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'Qtde_teorica' }).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_bing, left_on = 'Nome', right_on = 'Nome da Empresa', how = 'left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade Média (Anos)' : 'Idade' }).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ano_de_fundacao' : 'Idade' }).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais que 100 anos' if x > 100 else ('Menos que 50 anos' if x < 50 else 'Está entre 50 e 100'))
df_principal
```

```python
fig = px.bar(df_principal, x='Resultado', y='variacao_rs', text = 'variacao_rs', title='variação reais por resultado')
fig.show()
```
